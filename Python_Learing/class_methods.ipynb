```python
# class is a blueprint for creating instances 
#every employee we create will be an instances of that class 
class Employee:
    #add type hinting
    num_of_emps = 0
    raise_amount = 1.04 #class variable 


    def __init__(self, first,last, pay):# attributes 
        self.first = first #attributes 
        self.last = last 
        self.pay = pay
        self.email = first + last
#why not use self
        Employee.num_of_emps += 1 

    def full_name(self):
        #we add self for "full_name(self)" because if we run emp1.fullname(), without self, it will return an error that positional argument it is 0 but 1 was given because emp1 was the argument 
        return '{}{}'.format(self.first,self.last)
    #pass
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amount)

    #for def function to take class methods instead of regular, receive the class as the first argument instead of the instance 
    @classmethod 
    def set_raise_amt(cls,amount):
        # the variable is accepting amount from this method 
        cls.raise_amt = amount 

    @classmethod #alternative constructor
    #document step by step what this code is doing
    def from_string(cls,emp_str):
        first,last, pay = emp_str.split('-')
        return cls(first,last,pay)
    

    #should be static method if we do not access instance or class in the function 
    @staticmethod
    def is_workday(day):
        if day.weekday() == 5 or  day.weekday() == 6:
            return False
        return True 

emp1 = Employee('Ryan','Amy',5000) #instance of class 



Employee.set_raise_amt(1.06)


print(Employee.raise_amt)


emp_str_1 = "Jone-doe-2333"
emp_str_2 = "Alex-bliz-2345"
emp_str_3 = "Meg-rick-45333"

first,last, pay = emp_str_1.split('-')

new_emp_1 = Employee(first,last,pay)

new_emp_1.pay

import datetime 

my_date = datetime.date(2016,6,3)
print(Employee.is_workday(my_date))
```
